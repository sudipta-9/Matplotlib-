# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x = np.random.randint(10,60,(50))
print(x) #Give me some random data
```

```python
no = [10, 26, 35, 47, 57, 10, 24, 21, 34, 18, 24, 30, 25, 23, 39, 52, 51, 27, 56, 37, 24, 10, 36, 19,
 12, 41, 41, 38, 22, 29, 55, 50, 39, 41, 51, 46, 28, 30, 43, 22, 24, 57, 20, 16, 56, 25, 56, 52,
 55, 15]
plt.hist(no)
plt.title("Language")
plt.xlabel("Python")
plt.ylabel("Number")
plt.show() #Show the histogram graph
```

```python
plt.hist(no,color="r") #Change the color of red
```

```python
plt.hist(no,color="r",edgecolor="b") #Change edge color blue to clear vision
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b") #Set bins for clear xlabel dimension
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,"auto",(0,100),edgecolor="b") #Set the bins a perticular range
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",cumulative=-1) #Reverse the bar
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",bottom=10) #Set the bottom bar 10
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",align="mid")
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",histtype="step") #Change the histogram type
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",histtype="stepfilled")
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",histtype="barstacked")
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",histtype="bar")
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",orientation="horizontal") #Change the bar graph to horizontal
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",rwidth=0.8) #Change the width graph between them
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",log=True) #Change the ylabel value into logarithm
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",label="python")
plt.legend()
```

```python
l = [10,20,30,40,50,60]
plt.hist(no,color="r",bins=l,edgecolor="b",label="python")
plt.axvline(35,color="b",label="line")
plt.legend() #Add a line between 2 partition
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('language.csv')

print(data.columns)

column_name = 'Popularity'

plt.hist(data[column_name], bins=20, color='skyblue', edgecolor='black')
plt.title(f'Histogram of {column_name}')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
