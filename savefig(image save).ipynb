# Savefig

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line")
plt.show() #The graph is save in the particular folder
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line1",dpi=2000)
plt.show() #The save graph becomes more clear
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line2",facecolor="g")
plt.show() #Change the color of the boundry graph green 
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line3.pdf",facecolor="g")
plt.show() #Change the extension png to pdf. By default it is set to png
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line4",facecolor="g",transparent=True)
plt.show() #Make the full background color 
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="r")
plt.savefig("line5",facecolor="g",bbox_inches="tight")
plt.show()
```
