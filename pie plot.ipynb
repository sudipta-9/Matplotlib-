# Pie Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["c","c++","java","python"]
plt.pie(x,labels=y)
plt.show() #Show the basic pie graph
```

```python
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show() #Show a pie portion different way
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c)
plt.show() #Show the different colors
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True)
plt.show() #Show the percentage of each plot and there arise a shadow
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",radius=1)
plt.show() #Show increase and decrease of the pie graph
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",labeldistance=1.5)
plt.show() #Set the label near and far
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",startangle=90)
plt.show() #Rotate the circle
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",textprops={"fontsize":15,"color":"violet"})
plt.show() #Increase the font size and change font color
```

```python
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",counterclock=False)
plt.show() #Rotate clockwise
```

```python
c = ["r","b","g","y"]
plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%", shadow=True, wedgeprops={"linewidth":10,"edgecolor":"m"})
plt.show() #Enhance the shadow and edge color
```

```python
c = ["r","b","g","y"]
plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%", center=(2,3))
plt.show() #Change the center
```

```python
c = ["r","b","g","y"]
plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%", rotatelabels=True)
plt.show() #Rotate the label
```

```python
c = ["r","b","g","y"]
plt.pie(x, labels=y, explode=ex, colors=c, autopct="%0.1f%%", rotatelabels=True)
plt.title("Language")
plt.legend(loc=3)
plt.show() #Add the title and label with location
```

```python
plt.pie([1])
plt.show() #Dot pie chart
```

```python
x = [10,20,30,40]
z = [40,30,20,10]
y = ["c","c++","java","python"]
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(z,radius=1,colors=c)
plt.show() #There is a pie chart and under it another pie chart
```

```python
x = [10,20,30,40]
z = [40,30,20,10]
y = ["c","c++","java","python"]
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show() #Create a ring type pie chart
```

```python
x = [10,20,30,40]
z = [40,30,20,10]
y = ["c","c++","java","python"]
c = ["r","b","g","y"]
plt.pie(x,labels=y,radius=1.5)
cr = plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show() #Another way to create ring type pie chart
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("language.csv") 

labels = data['Language']    
values = data['Popularity'] 

plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Programming Language Popularity")
plt.legend(loc=1)
plt.show() #Import as a csv file and draw pie graph
```
