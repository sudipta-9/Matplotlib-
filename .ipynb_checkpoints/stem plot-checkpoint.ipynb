# Stem Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y)
plt.show() #Show the basic stem graph
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="+r")
plt.show() #Show the line dotted and make the marker + sign
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,markerfmt="ro",bottom=2)
plt.show() #The graph will start from bottom 2 of x axis
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,markerfmt="ro",bottom=6)
plt.show() #The graph will start from bottom 6 of y axis
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,markerfmt="ro",basefmt="g")
plt.show() #Change the bottom line color green
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,markerfmt="ro",basefmt="g",label="python")
plt.legend()
plt.show() #Show the label inside graph
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,markerfmt="ro",basefmt="g",label="python",orientation="horizontal")
plt.legend()
plt.show() #Change the orientation horizontal
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')  

x = data['Language']
y = data['Popularity']

plt.stem(x, y)

plt.xlabel('Language')
plt.ylabel('Popularity')
plt.title('Stem Plot')
plt.xticks(rotation=20)

plt.show() 
```
