# Bar Plot

```python
import matplotlib.pyplot as plt
x = ["python","java","c","javascript"]
y = [85,70,60,82]
plt.bar(x,y)
plt.show() #Show the sample bar plot
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y)
plt.show() #Set tile and name accross x and y axis
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r")
plt.show() #Change the width of the plot and color
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
c = ["y","b","m","g"]
plt.bar(x,y,width=0.5,color=c)
plt.show() #Set the different color for different bar
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",align="edge")
plt.show() #See the language name at the corner of each bar
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",align="center")
plt.show() #See the language name at the center of each bar. It is by default
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",edgecolor="g",linewidth=2)
plt.show() #Change the edge color of the bar graph and set width of line
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",edgecolor="g",linewidth=2,linestyle=":")
plt.show() #Change the line style to dot
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",alpha=0.5)
plt.show() #Set light color of the bar graph
```

```python
plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")
plt.bar(x,y,width=0.5,color="r",label="Popularity")
plt.legend()
plt.show() #Show a label with color inside bar plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","java","c","javascript"]
y = [85,70,60,82]
z = [15,20,25,30]

width=0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")

plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="y",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show() #Set the two different bar graph side wise with different color
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","java","c","javascript"]
y = [85,70,60,82]
z = [15,20,25,30]

width=0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=10)
plt.ylabel("number",fontsize=10)
plt.title("Sudipta")

plt.barh(p,y,width,color="r",label="Popularity")
plt.barh(p1,z,width,color="y",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show() #Set the horizontal bar graph
```
