# Text Operation

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph")
plt.show() #Show the simple title of the graph
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.show() #Change the size of the font
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.show() #Show the labels of x and y axis with perfect fontsize
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.text(2,5,"python")
plt.show() #There is a text with x-axis 2 and y-axis 5 inside the graph write python
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.text(2,5,"python",fontsize=15,style="italic")
plt.show() #Change the fontsize and style of the text inside the graph
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.text(2,5,"python",fontsize=15,bbox={"facecolor":"r"})
plt.show() #There arise a box around the text
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Python Graph",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Python",fontsize=15)
plt.annotate("python",xy=(2,1),xytext=(3,3),arrowprops=dict(facecolor="black",shrink=100))
plt.show() #There arise a arrow inside the graph of python
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]
plt.plot(x, y, label="Data Line")  
plt.title("Python Graph", fontsize=15)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Python", fontsize=15)
plt.legend(loc=0)  
plt.show() #There add a label inside the graph as a indicator and loc value should be 0-10
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]
plt.plot(x, y, label="Data Line")  
plt.title("Python Graph", fontsize=15)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Python", fontsize=15)
plt.legend(loc=0,facecolor="r",edgecolor="y",framealpha=0.5,shadow=True)  
plt.show() #Change the face and edge color of the label also light the label using framealpha as well as add shadow
```
