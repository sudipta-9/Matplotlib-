# Fill_Between Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.fill_between(x,y)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show() #Design a basic plot
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,color="r")
plt.show() #Set a range and design a box
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.fill_between(x,y,color="g")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show() #Fill with different color
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])
plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4))
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show() #There is a range of the fill color
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')

languages = data['Language']
popularity = data['Popularity']

x = range(len(languages))

plt.fill_between(x, popularity, color='skyblue', alpha=0.5)

plt.plot(x, popularity, marker='o', color='blue') 
plt.xticks(x, languages, rotation=45)
plt.title("Programming Language Popularity")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.grid(True)

plt.show() #By importing csv file draw fill between graph 
```
