# Box Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show() #Show the sample box plot
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,notch=True)
plt.show() #Change the shape of box
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,vert=False)
plt.show() #Box is in horizontal mode
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,widths=0.8)
plt.show() #Change the width of the box
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"])
plt.show() #At last show the label of python name
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,patch_artist=True)
plt.show() #Fill the color inside the box
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x,showmeans=True)
plt.show() #There is a indicator at the middle of box
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True)
plt.show() #Create a outler
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,whis=1.8)
plt.show() #It joins with the outler
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,sym="g+")
plt.show() #Change the outler circle to green color + sign
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,sym=" ")
plt.show() #It hides the outler
```

```python
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="g"),whiskerprops=dict(color="b"))
plt.show() #Change the box, caps and whisker color red, green and blue
```
