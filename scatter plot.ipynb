# Scatter Plot

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #See the basic scatter graph
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
sizes = [50,10,25,22,30,57,90]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #See the different color and size of the scatter plot
```

```python
plt.scatter(day,no,c=colors,s=sizes,alpha=0.5)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #Make the blur of the plot
```

```python
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=1)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #Change the circle maker to star and make color of the edge of plot and increase the width of the edge
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [25,65,10,59,60,80,45]
sizes = [50,10,25,22,30,57,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #See a color tool bar at the side of graph
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = [25,65,10,59,60,80,45]
sizes = [50,10,25,22,30,57,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
t = plt.colorbar()
t.set_label("ColorBar")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #See the colorbar title accross the color tool
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [6,5,8,4,2,3,1]
colors = [25,65,10,59,60,80,45]
sizes = [50,10,25,22,30,57,90]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.scatter(day,no2,color="r",s=sizes)
t = plt.colorbar()
t.set_label("ColorBar")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show() #See the two different plot at same area
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')  

plt.scatter(data['Language'], data['Popularity'])

plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.xticks(rotation=45) 
plt.show() #Draw the plot using csv file data import
```
