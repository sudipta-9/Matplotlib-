# Step Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.step(x,y)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show() #Design the simple step plot graph
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.step(x,y,marker="o")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show() #Show the marker with dot
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.step(x,y,color="r",marker="o",ms=10,mfc="g")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.show() #Change the line, marker color and marker size
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.step(x,y,color="r",marker="o",ms=10,mfc="g",label="python")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show() #Show the label
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')

languages = data['Language']
popularity = data['Popularity']

x = range(len(languages))

plt.step(x, popularity, where='mid')  

plt.xticks(x, languages, rotation=45)
plt.title("Programming Language Popularity")
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.grid(True)

plt.show() #Import csv file to show the step plot
```
