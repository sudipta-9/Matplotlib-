```python
import matplotlib.pyplot as plt
y = [20,50,40,30,50,40,30,70]
plt.plot(y)
plt.show() #Show the basic line graph
```

```python
import matplotlib.pyplot as plt
y = [20,50,40,30,50,40,30,70]
plt.plot(y,color="r",linestyle=":",linewidth=5)
plt.show() #Change the color of line and linestyle
```

```python
plt.plot(y,color="r",marker="o",mec="y")
plt.show() #Dot the marker at every point and color it
```

```python
plt.plot(y,color="r",alpha=0.5)
plt.show() #Reduce the transparency of graph
```

```python
import matplotlib.pyplot as plt
x = [40,20,10,40,30,60,80,40]
y = [20,50,40,30,50,40,30,70]
plt.plot(x)
plt.plot(y)
plt.show() #Show two different line within a same graph
```

```python
plt.plot(x,color="r",linewidth=2,marker="*")
plt.plot(y,color="g",linestyle=":",marker="o")
plt.legend(["men","women"])
plt.show() #Change the color, linewidth and linestyle of the lines
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')

plt.plot(data['Language'], data['Popularity'], marker='o', label='Data Points')
plt.title('Language Popularity')
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.xticks(rotation=20)
plt.grid(True)
plt.legend()
plt.show() #Import as a csv file
```
