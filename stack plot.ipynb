# Stack Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area = [2,3,2,5,4]
plt.stackplot(x,area)
plt.show() #Draw a simple stack plot
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l)
plt.legend(loc=2)
plt.show() #There are 3 different plots are presented with labels
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r","g","y"])
plt.legend(loc=2)
plt.show() #Change the colors
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r","g","y"], baseline="zero")
plt.legend(loc=2)
plt.show() #There is no changes
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r","g","y"], baseline="sym")
plt.legend(loc=2)
plt.show() #Remove the baseline
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r","g","y"], baseline="wiggle")
plt.legend(loc=2)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r","g","y"])
plt.legend(loc=2)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('language.csv')

languages = data['Language']
popularity = data['Popularity']

x = range(len(languages))

plt.stackplot(x, popularity, labels=languages, alpha=0.7)

plt.xticks(x, languages, rotation=45)
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Programming Language Popularity")
plt.legend(loc='upper right')

plt.show()
```
