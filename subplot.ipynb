```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('language.csv')

# Extract data
languages = data['Language']
popularity = data['Popularity']

# Create x-axis values
x = range(len(languages))

# Subplot 1 - Line plot
plt.subplot(2, 2, 1)
plt.plot(x, popularity, color="r")
plt.xticks(x, languages, rotation=45)
plt.title("Line Plot")

# Subplot 2 - Pie chart
plt.subplot(2, 2, 2)
plt.pie(popularity, labels=languages, autopct='%1.1f%%')
plt.title("Pie Chart")

# Subplot 3 - Stack plot
plt.subplot(2, 2, 3)
plt.stackplot(x, popularity, labels=languages, alpha=0.7)
plt.xticks(x, languages, rotation=45)
plt.title("Stack Plot")
plt.legend(loc='upper left')

# Subplot 4 - Bar chart
plt.subplot(2, 2, 4)
plt.bar(languages, popularity, color='green')
plt.xticks(x, languages, rotation=45)
plt.title("Bar Chart")

# Adjust subplot spacing
plt.tight_layout()
plt.show()

#Create a simple subplots by importing csv file
```
