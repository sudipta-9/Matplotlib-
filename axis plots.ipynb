# Axis Plots

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.xticks(x)
plt.show() #Change the value of x to whole number
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.yticks(y)
plt.show() #Change the value of y to whole number
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.xticks(x,labels=["python","java","c","c++","kotlin"])
plt.show() #Change the value of x number to letters
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show() #Set the limit value 0 to 10
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.axis ([0,10,0,7])
plt.show() #Set the value through x axis-0 to 10 and y axis-0 to 7
```
